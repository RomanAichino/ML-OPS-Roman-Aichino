```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('ML_data.csv')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

#aqu creamos una nube de palabras, donde veremos cuales fueron las mas utilizadas en los ttulos
#Cargar el archivo PNG del icono personalizado
icon_path = "imagen_film.png"
icon_image = Image.open(icon_path)

#Convertir la imagen en una matriz de NumPy
icon_array = np.array(icon_image)

#Crear una instancia de WordCloud con la mscara personalizada y otras opciones personalizadas
wordcloud = WordCloud(mask=icon_array, background_color='white').generate(' '.join(df['title'].tolist()))

#Mostrar la nube de palabras utilizando matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()

#la intencin era crearlo con una imagen de un icon de film, pero no salio </3
```

```python
#Cargar el archivo PNG del icono personalizado
icon_path = "imagen_film.png"
icon_image = Image.open(icon_path)

#Convertir la imagen en una matriz de NumPy
icon_array = np.array(icon_image)

#Crear una instancia de WordCloud con la mscara personalizada y otras opciones personalizadas
wordcloud = WordCloud(mask=icon_array, background_color='white').generate(' '.join(df['genres'].tolist()))

#Mostrar la nube de palabras utilizando matplotlib
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
df_ordenado = df.sort_values('original_language', ascending=True)
frecuencia= df_ordenado['original_language'].value_counts()
idiomas_relevantes = frecuencia[frecuencia >= 100].index.tolist()
df_filtrado = df[df['original_language'].isin(idiomas_relevantes)]
df_filtrado = df_filtrado.drop('Unnamed: 0', axis=1)
df_filtrado = df_filtrado.drop('genres', axis=1)
df_filtrado = df_filtrado.drop('title', axis=1)
df_filtrado = df_filtrado.drop('production_companies', axis=1)

sns.pairplot(df_filtrado, hue="original_language")
plt.show() #tarda alrededor de 25 segundos
```
